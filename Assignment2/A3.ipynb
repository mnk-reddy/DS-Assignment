```python
import numpy as np
import pandas as pan
import matplotlib.pyplot as mat
import sklearn as sky
```

```python
dataset=pan.read_csv('Data\BMI.csv')
dataset.head()
```

```python
X=dataset['Height']
Y=dataset['Weight']
mat.scatter(X,Y)
mat.show()
```

```python
from sklearn.preprocessing import normalize
data_scaled = normalize(dataset)
data_scaled = pan.DataFrame(data_scaled, columns=dataset.columns)
data_scaled.head()
```

```python
import scipy.cluster.hierarchy as shc
mat.figure(figsize=(10, 7))  
mat.title("Dendrograms")
dend = shc.dendrogram(shc.linkage(dataset))
mat.axhline(y=6, color='red', linestyle='-.')
```

```python
from sklearn.cluster import AgglomerativeClustering

H_clustering = AgglomerativeClustering(n_clusters=3)
H_clustering.fit(dataset)
predY = H_clustering.labels_
predY
```

```python
predY = H_clustering.labels_
colormap=np.array(['red','green','blue','yellow','orange','grey','violet'])
mat.scatter(dataset['Height'], dataset['Weight'], c=colormap[predY])
mat.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

H_clustering2 = AgglomerativeClustering(n_clusters=4)
H_clustering2.fit(dataset)
predY2 = H_clustering2.labels_
predY2
```
